```python
import cv2
import dlib 
import numpy
#
PREDICTOR_PATH = 'Data/shape_predictor_68_face_landmarks.dat'

predictor = dlib.shape_predictor(PREDICTOR_PATH)
detector = dlib.get_frontal_face_detector()
print(type(predictor))
#
print(type(detector))
```

```python
class TooManyFaces(Exception):
    pass

class NoFaces(Exception):
    pass

def get_landmarks(im):
    rects = detector(im, 1)

    if len(rects) > 1:
        raise TooManyFaces
    if len(rects) == 0:
        raise NoFaces

    return numpy.matrix([[p.x, p.y] for p in predictor(im, rects[0]).parts()])

def annotate_landmarks(im, landmarks):
    im = im.copy()
    for idx, point in enumerate(landmarks):
        pos = (point[0, 0], point[0, 1])
        print(pos)
        cv2.putText(im, str(idx), pos,
                    fontFace=cv2.FONT_HERSHEY_SCRIPT_SIMPLEX,
                    fontScale=0.4,
                    
                    color=(0, 0, 255))
        cv2.circle(im, pos, 3, color=(0, 255, 255))
    return im
```

```python
image = cv2.imread('images/kim1.jpg')
landmarks = get_landmarks(image)
image_with_landmarks = annotate_landmarks(image, landmarks)   
#
#print(landmarks) ## Ok Dont --- its a 
print(len(landmarks))


print(landmarks)


print(type(landmarks)) ### <class 'numpy.matrixlib.defmatrix.matrix'>
## Source --- Same as Numpy Matrix ?? Probably NOT 
## Some kind of FLATTEN MATRIX 
## https://github.com/numpy/numpy/issues/5269
## https://docs.scipy.org/doc/numpy/reference/generated/numpy.matrix.flatten.html
##### https://www.pythonlearn.com/html-008/cfbook011.html
## https://stackoverflow.com/questions/41031186/whats-the-differences-between-a-numpy-matrix-and-a-numpy-matrixlib-defmatrix-ma

```
